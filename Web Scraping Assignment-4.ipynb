```python
for i in driver.find_elements(By.XPATH,'//a[@class="Link"]'):
    Repository_title.append(i.text)
print(Repository_title,len(Repository_title))
```

```python
for i in driver.find_elements(By.XPATH,'//h2[@class="h3 lh-condensed"]/a'):
    Url.append(i.get_attribute('href'))
print(Url,len(Url))
```

```python
driver=webdriver.Chrome()
driver.get('https://archive.ics.uci.edu/')
```

```python
Dataset_name=[]
Data_type=[]
Task=[]
Attribute_type=[]
No_of_instances=[]
No_of_attribute=[]
```

```python
view_dataset=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div/div/div/a[1]').click()
```

```python
#Since there are 664 dataset and there are 25 dataset per page so we will have to iterate 27 times
```

```python
for i in range(26):
    for i in driver.find_elements(By.XPATH,'//div[@class="relative col-span-8 sm:col-span-7"]/h2/a'):
        Dataset_name.append(i.text)
    for i in driver.find_elements(By.XPATH,'//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]/div[1]/span'):
        Data_type.append(i.text)
    for i in driver.find_elements(By.XPATH,'//div[@class="relative col-span-8 sm:col-span-7"]/p'):
        Task.append(i.text)
    for i in driver.find_elements(By.XPATH,'//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]/div[2]/span'):
        Attribute_type.append(i.text)
    for i in driver.find_elements(By.XPATH,'//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]/div[3]/span'):
        No_of_instances.append(i.text)
    for i in driver.find_elements(By.XPATH,'//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]/div[4]/span[1]'):
        No_of_attribute.append(i.text)
    next_button=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[3]/div/button[2]')
    next_button.click()
driver.close()
```

```python
df=pd.DataFrame({'Dataset_name':Dataset_name,'Data_type':Data_type,'Task':Task,'Attribute_type':Attribute_type,'No_of_instances':No_of_instances,'No_of_attributes':No_of_attribute})
```
